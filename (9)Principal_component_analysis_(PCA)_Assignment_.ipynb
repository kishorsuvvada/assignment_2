###  Dimensionality Reduction with PCA

Principal Component Analysis (PCA) is a widely used technique for dimensionality reduction that transforms a large set of correlated variables into a smaller set of uncorrelated variables called principal components.

```python
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
```

```python
wine_norm=scale(wine)
```

```python
wine_norm
```

```python
pca=PCA()
pca_val=pca.fit_transform(wine_norm)
```

```python
pca_val
```

```python
varaible=pca.explained_variance_ratio_
varaible
```

```python
var1 = np.cumsum(np.round(varaible,decimals = 4)*100)
var1
```

```python
pca.components_
```

### Clustering with PCA Data

After reducing the dimensionality of the dataset using PCA, you can apply a clustering algorithm (like K-means) on the PCA-transformed data. Clustering on the reduced data can often help with better performance and more meaningful results, especially when the dataset has high dimensionality.

```python
from sklearn.cluster import KMeans
```

```python
# K-means for pca data
model=KMeans(n_clusters=6).fit(pca_data)
```

```python
pca_data.plot(x="X",y="Y",c=model.labels_,kind="scatter",s=50,cmap=plt.cm.coolwarm_r)
```

```python
import sklearn
sklearn.metrics.silhouette_score(pca_data,model.labels_)
```

### Clustering with Original Data

Before applying PCA, we should first attempt clustering on the original data to serve as a comparison point. This involves applying a clustering algorithm such as K-means, visualizing the results, and evaluating the clustering performance using appropriate metrics.

```python
import warnings
warnings.filterwarnings("ignore")
fig=plt.figure(figsize=(8,4))
WCSS=[]
for i in range(1,11):
    model=KMeans(n_clusters=i).fit(wine_norm)
    WCSS.append(model.inertia_)

plt.plot(range(1,11),WCSS)
plt.title("The Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
# Silhouette Score Exploration
from sklearn.metrics import silhouette_score

silhouette_scores = []
components_range = range(2, 11)

for n_components in components_range:
    pca = PCA(n_components=n_components)
    pca_values_n = pca.fit_transform(wine_norm)
    score = silhouette_score(pca_values_n, model.labels_)
    silhouette_scores.append(score)

plt.figure(figsize=(10, 5))
plt.plot(components_range, silhouette_scores, marker='o', linestyle='--', color='blue')
plt.title('Silhouette Score vs Number of PCA Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
# Experiment with different numbers of principal components
for n_components in [2, 3, 5, 10]:
    pca = PCA(n_components=n_components)
    pca_values_n = pca.fit_transform(wine_norm)

    plt.figure(figsize=(10, 5))
    plt.scatter(pca_values_n[:, 0], pca_values_n[:, 1], c=model.labels_, cmap='viridis', s=50)
    plt.title(f'PCA with {n_components} Components')
    plt.xlabel('PCA 1')
    plt.ylabel('PCA 2')
    plt.colorbar(label='Cluster')
    plt.show()
```

# **Conclusion** **and Insights**

PCA improves the performance of clustering: The value of the silhouette score increased substantially after the application of PCA, and this, therefore, points out that by performing dimensionality reduction, better-defined cluster separation can be exposed.

Dimensionality reduction trade-offs: While PCA simplifies the data, it actually retains sufficient variance to improve the performance of clustering, particularly when the high-dimensional nature of the data may obscure underlying patterns in some instances.
